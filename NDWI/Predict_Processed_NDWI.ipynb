```python
import os
from itertools import product

import numpy as np
from PIL import Image

import tensorflow as tf
from keras.preprocessing import image
```

```python
image_path = "/content/drive/MyDrive/3rd Year Project/NDWI/Copy of LC08_L2SP_141045_20210516_20210525_02_T1_SR_nir.png"
cropped_image_path = "/content/Crop"
model_path = "/content/drive/MyDrive/3rd Year Project/NDWI/Model/NDWI_Rivers_Lakes_Trained_Model.h5"
```

```python
dh = 1024
dw = 1024
img = Image.open(image_path)
w, h = img.size

grid = product(range(0, h-(h % dh), dh), range(0, w-(w % dw), dw))
for i, j in grid:
    if j+dw <= w:
        box = (j, i, j+dw, i+dh)
        out = os.path.join(cropped_image_path, f'{j}x{i}_{j+dw}x{i+dh}.png')
        cropped_img = img.crop(box)
        np_cropped_img = np.asarray(cropped_img)
        if np.all(np_cropped_img == np_cropped_img[0]):
            continue
        else:
            cropped_img.save(out)
```

```python
model = tf.keras.models.load_model(model_path)

class_names = ["Lake", "River"]
```

```python
img_width, img_height = 1024, 1024

for img_name in os.listdir(cropped_image_path):
    img = image.load_img(os.path.join(cropped_image_path, img_name))
    img = image.img_to_array(img)
    img = np.expand_dims(img, axis = 0)

    predictions = model.predict(img)

    score = tf.nn.softmax(predictions[0])

    print(
        "{} image most likely belongs to {} with a {:.2f} percent confidence."
        .format(img_name, class_names[np.argmax(score)], 100 * np.max(score))
    )
```
